input of files and simple modifications of dates.

```python
import pandas as pd
df=pd.read_csv("CORPORATE_ACTIONS/nse_dividends.csv")
df['announcement_date'] = pd.to_datetime(df['announcement_date'], dayfirst=True, format='%d-%b-%Y')
df['announcement_date'] = df['announcement_date'].dt.strftime('%Y-%m-%d')
df['record_date'] = pd.to_datetime(df['record_date'], dayfirst=True, format='%d-%b-%Y')
df['record_date'] = df['record_date'].dt.strftime('%Y-%m-%d')
df['pay_date'] = pd.to_datetime(df['pay_date'], dayfirst=True, format='%d-%b-%Y')
df['pay_date'] = df['pay_date'].dt.strftime('%Y-%m-%d')

df
```
