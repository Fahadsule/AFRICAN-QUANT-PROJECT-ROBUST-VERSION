We are just gonna plot this graph from a book

```python

import pandas as pd
x_values=[0,5,10,15,20,25,30]
y_values=[1000,694,444,250,111,28,0]
df=pd.DataFrame({
    'time':x_values,
    'volume':y_values
})
df
```

NOW LETS GRAPH THIS 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df['time'], df['volume'], marker='o', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('time', fontsize=12)
plt.ylabel('volume', fontsize=12)
plt.title('Line Plot of volume vs time', fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
```

LETS CREATE A SYSTEM THAT PLOTS FUNCTIONS.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = np.arange(-1000, 1000.001, 0.001)
y=(x**2)
print(y)
df=pd.DataFrame({'x':x, 'y':y})
import matplotlib.pyplot as plt
plt.figure(figsize=(40, 20))
plt.plot(df['x'], df['y'], marker='o', linestyle='-', linewidth=0.1, markersize=0.1)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Line Plot of y vs x', fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
```
